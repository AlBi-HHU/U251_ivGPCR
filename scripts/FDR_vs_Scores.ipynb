```python
sb.set_style("whitegrid")
data = pd.read_csv("UL33IREFKEGG.csv")
#data["FDR"]=data["FDR"].iloc[::-1]
#ax.set_size_inches(11.7, 8.27)
sb.set(font_scale = 1.4)
lm = sb.boxplot(x = "FDR", y = "Score", data = data, fliersize=8)
ax = sb.boxplot(x=data["FDR"], y=data["Score"],showmeans=True)
ax.invert_xaxis()

# Calculate number of obs per group & median to position labels
medians = data.groupby(['FDR'])['Score'].median().values
nobs = data['FDR'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
nobs = nobs[::-1]

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick], 1.01, nobs[tick],
    horizontalalignment='center', size='small', color='b', weight='semibold')
    
fig = ax.get_figure()
fig.set_size_inches(11.7, 8.27)
fig.savefig("UL33IREFKEGG.png") 
```
